```python
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df['Score'] = df['Score'].fillna(df['Score'].median())
df
```

```python
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
df['Gross'] = df['Gross'].fillna(df['Gross'].mean())
df['Gross']
```

```python
df['Cast3'] = df['Cast3'].fillna(df['Cast3'].mode()[0])
df['Cast4'] = df['Cast4'].fillna(df['Cast4'].mode()[0])
df
```

```python
# Rating × votes ilə “IMDB_score_power” adlı yeni sütun yaradın (yüksək reytinq * çox səs = daha güclü film).
df['IMDb_score_power']=df['Rating']*df['Votes']
df
```

```python
# Datasetdə təkrarlanan sətir (duplicate) olub-olmadığını yoxlayın və varsa silin.
df.duplicated().sum() #dublicat deyer yoxdur
#df.drop_duplicates(inplace=True)
```

```python
# Directora görə qruplaşdırın və hər bir rejissorun neçə film çəkdiyini tapın.
# → Ən çox filmi olan ilk 5 rejissoru göstər.
# Rejissorlara görə qruplaşdırıb film sayını tapmaq
director_counts = df.groupby('Director')['Title'].count().reset_index()
# Sütunun adını dəyişddim
director_counts.rename(columns={'Title':'Film_Count'}, inplace=True)
# Ən çox filmi olan ilk 5 rejissor
top5_directors = director_counts.sort_values(by='Film_Count', ascending=False).head(5)
print(top5_directors)
```
